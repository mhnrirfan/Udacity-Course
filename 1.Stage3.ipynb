# Sigma Labs: Advised Knowledge for Stage 3

Welcome to your Stage 3 Interview. This document will provide you with the knowledge you need to be successful in the interview. You should be familiar with the topics listed below. If you are not, please take the time to review them before your interview.

## General

- [ ] You should be able to:
    - [ ] Understand how indentation affects code in Python
    - [ ] Print to the console using `print()`
    - [ ] Use Python comments to write non-code in Python files
    - Some useful built-in functions are:
        - [ ] `input()`: Used to get input from the user. For example, `input("Enter your name: ")` will prompt the user to enter their name.
        - [ ] `print()`: Used to print output to the console. For example, `print("Hello, World!")` will print "Hello, World!" to the console.
        - [ ] `type()`: Used to get the type of an object. For example, `type(5)` will return `<class 'int'>`

## Variables

- [ ] You should be able to:
    - [ ] Declare a variable
    - [ ] Retrieve data from a variable
    - [ ] Change the value of a variable after it had been set
    - [ ] Store the output of a function in a variable

- For example, you should be able to:
    - [ ] Declare a variable called `name` and set it to `"John"`
    - [ ] Print the value of the variable `name` to the console
    - [ ] Change the value of the variable `name` to `"Jane"`
    - [ ] Declare a variable called `numbers` and set it to `[1, 2, 3]`

## Functions

- [ ] You should be able to:
    - [ ] Define a function
    - [ ] Invoke (or Call) a function so that it executes
    - [ ] Describe the role of parameters in a function
    - [ ] Describe the role of `return` in a function
    - [ ] Describe what the difference between `return` and `print()` are in a function

- For example, you should be able to:
    - [ ] Define a function that adds two numbers together and returns them
    - [ ] Define a function that takes a list of numbers and returns the sum of all the numbers in the list

## If Statements

- [ ] You should be able to:
    - [ ] Understand what an `if` statement is
    - [ ] Write an `if` statement that checks if a condition is true
    - [ ] Write an `if` statement that checks if a condition is false
    - [ ] Write an `if` statement that checks multiple conditions
    - [ ] Write an `if` statement that uses logical operators (e.g., `and`, `or`, `not`)
    - [ ] Write an `if` statement that uses comparison operators (e.g., `==`, `!=`, `<`, `>`, `<=`, `>=`)
    - [ ] Use an `else` statement to execute code when an `if` statement is false
    - [ ] Use an `elif` statement to check multiple conditions

- Some useful built-in functions are:
    - [ ] `isinstance()`: Used to check if an object is an instance of a class. For example, `isinstance(5, int)` will return `True`
    - [ ] `type()`: Used to get the type of an object. For example, `type(5)` will return `<class 'int'>`

## Loops

- [ ] You should be able to:
    - [ ] Understand what a loop is
    - [ ] Write a `for` loop that iterates over a list
    - [ ] Write a `for` loop that iterates over a list using its index

- Some useful built-in functions are:
    - [ ] `range()`: Used to generate a sequence of numbers. For example, `range(5)` will return `[0, 1, 2, 3, 4]`. In the context of a `for` loop you could use it like this: `for i in range(5):`, which will then count up through `i`.
    - [ ] `enumerate()`: Used to get the index and value of elements in a list. For example, `enumerate(["a", "b", "c"])` will return `[(0, "a"), (1, "b"), (2, "c")]`. In the context of a `for` loop, you could use it like this: `for i, value in enumerate(["a", "b", "c"]):`, which will then give you the index and value of the list.

## Strings

- [ ] You should be able to:
    - [ ] Store strings in variables
    - [ ] Concatenate strings (e.g., `my_string = "Hello" + " " + "World"`)
    - [ ] Access characters in a string using indexing (e.g., `my_string[0]` to get the first letter of the string)
    - [ ] Use string slicing to extract a substring from a string (e.g., `my_string[0:5]` to get the first five letters of the string)

- Some useful built-in functions are:
    - [ ] `split()`: Used to convert a string into a list of strings. For example, `"a.b.c".split(".")` will return `["a", "b", "c"]`
    - [ ] `join()`: Used to convert a list of strings into a single string. For example, `".".join(["a", "b", "c"])` will return `"a.b.c"`
    - [ ] `index()`: Used to find the index of a character in a string. For example, `"abc".index("b")` will return `1`
    - [ ] `replace()`: Used to replace a character in a string with another character. For example, `"abc".replace("b", "d")` will return `"adc"`
    - [ ] `upper()`: Used to convert a string to uppercase. For example, `"abc".upper()` will return `"ABC"`
    - [ ] `lower()`: Used to convert a string to lowercase. For example, `"ABC".lower()` will return `"abc"`
    - [ ] `strip()`: Used to remove whitespace from the beginning and end of a string. For example, `" abc ".strip()` will return `"abc"`
    - [ ] `lstrip()`: Used to remove whitespace from the beginning of a string. For example, `" abc ".lstrip()` will return `"abc "`
    - [ ] `rstrip()`: Used to remove whitespace from the end of a string. For example, `" abc ".rstrip()` will return `" abc"`
    - [ ] `startswith()`: Used to check if a string starts with a certain character. For example, `"abc".startswith("a")` will return `True`
    - [ ] `endswith()`: Used to check if a string ends with a certain character. For example, `"abc".endswith("c")` will return `True`

## Numbers

- [ ] You should be able to:
    - [ ] Store numbers in variables
    - [ ] Perform basic arithmetic operations (e.g., `+`, `-`, `*`, `/`)
    - [ ] Use comparison operators (e.g., `==`, `!=`, `<`, `>`, `<=`, `>=`)

- Some useful built-in functions are:
    - [ ] `abs()`: Used to return the absolute value of a number. For example, `abs(-5)` will return `5`
    - [ ] `round()`: Used to round a number to the nearest whole number. For example, `round(3.7)` will return `4`
    - [ ] `int()`: Used to convert a number to an integer. For example, `int(3.7)` will return `3`
    - [ ] `float()`: Used to convert a number to a float. For example, `float(3)` will return `3.0`

## Booleans

- [ ] You should be able to:
    - [ ] Understand what a boolean is
    - [ ] Use boolean operators (e.g., `and`, `or`, `not`)
    - [ ] Use comparison operators (e.g., `==`, `!=`, `<`, `>`, `<=`, `>=`)

- Some useful built-in functions are:
    - [ ] `bool()`: Used to convert a value to a boolean. For example, `bool(0)` will return `False`
    - [ ] `all()`: Used to check if all elements in an iterable are true. For example, `all([True, True, False])` will return `False`

## Lists

- [ ] You should be able to:
    - [ ] Create a list (e.g., `my_list = [1, 2, 3]`)
    - [ ] Access elements in a list (e.g., `first_element = my_list[0]`)
    - [ ] Modify elements in a list (e.g., `my_list[0] = 4`)
    - [ ] Add elements to a list (e.g., `my_list.append(4)`)
    - [ ] Remove elements from a list (e.g., `my_list.remove(4)`)

- Some useful built-in functions are:
    - [ ] `len()`: Used to get the length of a list. For example, `len([1, 2, 3])` will return `3`
    - [ ] `max()`: Used to get the maximum value in a list. For example, `max([1, 2, 3])` will return `3`
    - [ ] `min()`: Used to get the minimum value in a list. For example, `min([1, 2, 3])` will return `1`
    - [ ] `sum()`: Used to get the sum of all elements in a list. For example, `sum([1, 2, 3])` will return `6`
    - [ ] `sort()`: Used to sort a list in ascending order. For example, `[3, 1, 2].sort()` will return `[1, 2, 3]`
    - [ ] `reverse()`: Used to reverse the order of elements in a list. For example, `[1, 2, 3].reverse()` will return `[3, 2, 1]`
    - [ ] `count()`: Used to count the number of occurrences of an element in a list. For example, `[1, 2, 2, 3].count(2)` will return `2`
    - [ ] `index()`: Used to find the index of an element in a list. For example, `[1, 2, 3].index(2)` will return `1`

## Dictionaries

- [ ] You should be able to:
    - [ ] Define what a dictionary is and what it is used for
    - [ ] Create a dictionary (e.g., `my_dict = {"name": "John", "age": 30}`)
    - [ ] Access values in a dictionary (e.g., `name = my_dict["name"]`)
    - [ ] Modify values in a dictionary (e.g., `my_dict["age"] = 31`)
    - [ ] Add key-value pairs to a dictionary
    - [ ] Remove key-value pairs from a dictionary
    - [ ] Check if a key exists in a dictionary

- Some useful built-in functions are:
    - [ ] `len()`: Used to get the number of key-value pairs in a dictionary. For example, `len({"name": "John", "age": 30})` will return `2`
    - [ ] `keys()`: Used to get a list of all keys in a dictionary. For example, `{"name": "John", "age": 30}.keys()` will return `["name", "age"]`
    - [ ] `values()`: Used to get a list of all values in a dictionary. For example, `{"name": "John", "age": 30}.values()` will return `["John", 30]`
    - [ ] `items()`: Used to get a list of all key-value pairs in a dictionary. For example, `{"name": "John", "age": 30}.items()` will return `[("name", "John"), ("age", 30)]`
    - [ ] `get()`: Used to get the value of a key in a dictionary. For example, `{"name": "John", "age": 30}.get("name")` will return `"John"`
